### Tuning DBSCAN Parameters

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score

best_eps = 0
best_min_samples = 0
highest_silhouette_score = -1

# Define ranges for eps and min_samples
eps_values = np.arange(0.1, 2.0, 0.1)
min_samples_values = range(2, 20)

for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        clusters = dbscan.fit_predict(pca_data)

        # Exclude noise points (-1) for silhouette score calculation
        valid_clusters_mask = clusters != -1
        if sum(valid_clusters_mask) > 1 and len(set(clusters[valid_clusters_mask])) > 1:
            score = silhouette_score(pca_data[valid_clusters_mask], clusters[valid_clusters_mask])
            if score > highest_silhouette_score:
                highest_silhouette_score = score
                best_eps = eps
                best_min_samples = min_samples

print(f"Optimal eps: {best_eps}")
print(f"Optimal min_samples: {best_min_samples}")
print(f"Highest Silhouette Score: {highest_silhouette_score}")
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import pandas as pd
import numpy as np

# df_no_outliers and pca_data_no_outliers within this cell to resolve NameError
# This part is reconstructed from the previously deleted steps for outlier removal and PCA
is_not_outlier = pd.Series(True, index=df_model.index)

for column in df_model.columns:
    Q1 = df_model[column].quantile(0.25)
    Q3 = df_model[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    is_not_outlier = is_not_outlier & (
        (df_model[column] >= lower_bound) & (df_model[column] <= upper_bound)
    )

df_no_outliers = df_model[is_not_outlier].copy()

scaler_no_outliers = StandardScaler()
scaled_data_no_outliers = scaler_no_outliers.fit_transform(df_no_outliers)

pca_no_outliers = PCA(n_components=2)
pca_data_no_outliers = pca_no_outliers.fit_transform(scaled_data_no_outliers)


best_eps = 0
best_min_samples = 0
highest_silhouette_score = -1

# Define ranges for eps and min_samples
eps_values = np.arange(0.1, 2.0, 0.1)
min_samples_values = range(2, 20)

for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        clusters = dbscan.fit_predict(pca_data_no_outliers)

        # Exclude noise points (-1) for silhouette score calculation
        valid_clusters_mask = clusters != -1
        # Check if there are at least 2 clusters (excluding noise) and more than 1 sample
        if sum(valid_clusters_mask) > 1 and len(set(clusters[valid_clusters_mask])) > 1:
            score = silhouette_score(pca_data_no_outliers[valid_clusters_mask], clusters[valid_clusters_mask])
            if score > highest_silhouette_score:
                highest_silhouette_score = score
                best_eps = eps
                best_min_samples = min_samples

print(f"Optimal eps: {best_eps}")
print(f"Optimal min_samples: {best_min_samples}")
print(f"Highest Silhouette Score: {highest_silhouette_score}")
```

```python
dbscan_optimal_no_outliers = DBSCAN(eps=best_eps, min_samples=best_min_samples)
df_no_outliers['DBSCAN_No_Outliers_Cluster'] = dbscan_optimal_no_outliers.fit_predict(pca_data_no_outliers)

valid_clusters_mask_no_outliers = df_no_outliers['DBSCAN_No_Outliers_Cluster'] != -1
if sum(valid_clusters_mask_no_outliers) > 1 and len(set(df_no_outliers['DBSCAN_No_Outliers_Cluster'][valid_clusters_mask_no_outliers])) > 1:
    sil_score_dbscan_optimal_pca_no_outliers = silhouette_score(
        pca_data_no_outliers[valid_clusters_mask_no_outliers],
        df_no_outliers['DBSCAN_No_Outliers_Cluster'][valid_clusters_mask_no_outliers]
    )
    print(f"DBSCAN (PCA, no outliers, tuned) Silhouette Score: {sil_score_dbscan_optimal_pca_no_outliers}")
else:
    print("DBSCAN (PCA, no outliers, tuned) Silhouette Score: Cannot compute (less than 2 clusters or all noise)")

plt.figure(figsize=(8,6))
plt.scatter(
    pca_data_no_outliers[:, 0],
    pca_data_no_outliers[:, 1],
    c=df_no_outliers['DBSCAN_No_Outliers_Cluster']
)
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("DBSCAN Clusters (PCA View) - Outliers Removed, Tuned Parameters")
plt.show()
```
